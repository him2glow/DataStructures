reverse a Queue using data swapping

```
class Queue():
    def __init__(self,size):
        self.front = self.rear=-1
        self.size = size
        self.queue = [None for i in range(self.size)]

    def Enque(self,data):
        if (self.rear + 1) % self.size == self.front:
            print("Queue is full")
        elif self.front==-1:
            self.front=0
            self.rear=0
            self.queue[self.rear]=data
        else:
            self.rear = (self.rear+1)%self.size
            self.queue[self.rear]=data

    def reverseQ(self):
        while self.front<self.rear:
            self.queue[self.front],self.queue[self.rear]=self.queue[self.rear],self.queue[self.front]
            self.rear-=1
            self.front+=1

    def display(self):

        # condition for empty queue
        if (self.front == -1):
            print("Queue is Empty")

        else:
            print("Elements in the circular queue are1:", end=" ")
            for i in range(0,self.size):
                print(self.queue[i], end=" ")
            print()






q = Queue(5)
q.Enque(1)
q.Enque(2)
q.Enque(3)
q.Enque(4)
q.Enque(5)
q.display()
q.reverseQ()
q.display()


```

## Queue using Stacks 

enQueue(q, x):
While stack1 is not empty, push everything from stack1 to stack2.
Push x to stack1 (assuming size of stacks is unlimited).
Push everything back to stack1.
Here time complexity will be O(n)

deQueue(q):
If stack1 is empty then error
Pop an item from stack1 and return it
Here time complexity will be O(1)

```
#by making enque operation Costly
#enque = O(n)
#deque = O(1)

class Queue():
    def __init__(self):
        self.s1 = []
        self.s2 = []

    def enqueue(self,data):
        while len(self.s1)!=0:
            self.s2.append(self.s1.pop())
        self.s1.append(data)

        while len(self.s2)!=0:
            self.s1.append(self.s2.pop())


    def deQueue(self):
        if len(self.s1)==0:
            print("list is empty")
        x = self.s1.pop()
        print("dequued element is:",x)


    def queueFront(self):
        if len(self.s1)==0:
            print("list is empty")
        x = self.s1[-1]
        print("front element is:",x)

    def display(self):
        if len(self.s1)==0:
            print("list is empty")
        while len(self.s1)!=0:
            x = self.s1.pop()
            print(x,end=" ")





q = Queue()
q.enqueue(1)
q.enqueue(2)
q.enqueue(31)
q.enqueue(3)
#q.deQueue()
#q.deQueue()
q.queueFront()
q.display()

```

Circular Queue Implewmentation

```
class CircularQueue():

    # constructor
    def __init__(self, size):  # initializing the class
        self.size = size

        # initializing queue with none
        self.queue = [None for i in range(size)]
        self.front = self.rear = -1

    def enqueue(self, data):

        # condition if queue is full
        if ((self.rear + 1) % self.size == self.front):
            print(" Queue is Full\n")

        # condition for empty queue
        elif (self.front == -1):
            self.front = 0
            self.rear = 0
            self.queue[self.rear] = data
        else:

            # next position of rear
            self.rear = (self.rear + 1) % self.size
            self.queue[self.rear] = data

    def dequeue(self):
        if (self.front == -1):  # codition for empty queue
            print("Queue is Empty\n")

        # condition for only one element
        elif (self.front == self.rear):
            temp = self.queue[self.front]
            self.front = -1
            self.rear = -1
            return temp
        else:
            temp = self.queue[self.front]
            self.front = (self.front + 1) % self.size
            return temp

    def display(self):

        # condition for empty queue
        if (self.front == -1):
            print("Queue is Empty")

        elif (self.rear >= self.front):
            print("r",self.rear)
            print("f",self.front)

            print("Elements in the circular queue are1:",end=" ")
            for i in range(self.front, self.rear + 1):
                print(self.queue[i], end=" ")
            print()

        else:
            #once the list if full and you delete some element and insert again...rear will again start from 0 
            #since it is a circular queue
            #that is when you will need this loop
            print("Elements in Circular Queue are2:", end=" ")
            for i in range(self.front, self.size):
                print(self.queue[i], end=" ")
            for i in range(0, self.rear + 1):
                print(self.queue[i], end=" ")
            print()


        if ((self.rear + 1) % self.size == self.front):
            print("Queue is Full")

# Driver Code
ob = CircularQueue(5)
ob.enqueue(1)
ob.enqueue(2)
ob.enqueue(3)
ob.enqueue(4)
ob.enqueue(5)
ob.display()
print("Deleted value = ", ob.dequeue())
print("Deleted value = ", ob.dequeue())
ob.display()
ob.enqueue(9)
ob.enqueue(20)
ob.display()

# This code is contributed by AshwinGoel
```

Circular Queue using Linked List

```
class Node():
    def __init__(self,data):
        self.item = data
        self.next=None
class Queue():
    def __init__(self):
        self.front = None
        self.rear = None

    def enque(self,data):
        new_node = Node(data)
        #if list is empty
        if self.front==None:
            self.front = new_node
        else:
            self.rear.next=new_node
        self.rear = new_node
        self.rear.next=self.front


    def display_Queue(self):
        if self.front==None:
            print("Queue is empty")
        else:
            temp = self.front
            while temp.next != self.front:
                print(temp.item,end=' ')
                temp = temp.next

            print(temp.item)


    def DEqueue(self):
        if self.front==None:
            print("empty Queue")
        #for only 1 element
        value = None
        if self.front==self.rear:
            value = self.front.item
            self.front = None
            self.rear = None
        else:
            value = self.front.item
            self.front = self.front.next
            self.rear.next = self.front
        return value



Q = Queue()
Q.enque(1)
Q.enque(2)
Q.enque(3)
Q.enque(4)
print(Q.DEqueue())
print(Q.DEqueue())


Q.display_Queue()
```

Priority Queue using Linked LIst

```
class Node():
    def __init__(self,data,priority):
        # three fields
        self.item = data
        self.pri = priority
        self.next=None

class PriorityQueue():
    def __init__(self):
        self.head = None

    def enque(self,data,priority):
        new_node  = Node(data,priority)
        if self.head is None:
            self.head=new_node
        elif new_node.pri<self.head.pri:
            new_node.next=self.head
            self.head=new_node
        else:
            temp=self.head
            while temp.next and temp.next.pri<=new_node.pri:
                temp = temp.next
            new_node.next = temp.next
            temp.next = new_node



    def display_Queue(self):
        #just need to return the first element
        if self.head is None:
            print("Priority QUEUE is empty")
            return
        else:
            temp  = self.head
            while temp:
                print(temp.item,end=" ")
                temp=temp.next

    def DEqueue(self):
        if self.head is None:
            print("Empty Queue")
            return
        temp = self.head.item
        self.head=self.head.next
        return temp




Q = PriorityQueue()
Q.enque(4,1)
Q.enque(5,2)
Q.enque(15,2)
Q.enque(6,4)
Q.enque(7,0)
print(Q.DEqueue())
print(Q.DEqueue())
Q.display_Queue()
```
